## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
import pandas as pd

# Load dataset and construct the binary target column
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')
df['is_declining'] = (df['trend_direction'] == 'down').astype(int)

print("Target Column ('is_declining') Distribution:")
print(df['is_declining'].value_counts(normalize=True).map(lambda x: f"{x:.2%}"))
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# Baseline precision check (random guessing threshold)
baseline_precision = df['is_declining'].mean()
print(f"Baseline Precision (Random Guessing): {baseline_precision:.2%}")
print(f"Target Success Threshold: Model Precision@100 > {baseline_precision * 1.5:.2%}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# Demonstrating why a simple rule (e.g., Age > 365 days) fails
old_pages = df[df['content_age_days'] > 365]
rule_precision = (old_pages['trend_direction'] == 'down').mean()

print(f"Precision of hard-coded rule (IF age > 365 days): {rule_precision:.2%}")
print("Conclusion: A simple age rule catches many non-declining pages. ML is needed to combine non-linear signals accurately.")
```
