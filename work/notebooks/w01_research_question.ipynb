```python
import pandas as pd

# Load the starter dataset from the repository (Colab will find it now!)
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

# 1. Total number of pages available for our analysis
total_pages = len(df)

# 2. Let's see how many of these pages are currently losing traffic
declining_pages = len(df[df['trend_direction'] == 'down'])
decline_rate = (declining_pages / total_pages) * 100

# 3. Let's pull a quick signal test: does age correlate with decline?
avg_age_down = df[df['trend_direction'] == 'down']['content_age_days'].mean()
avg_age_other = df[df['trend_direction'] != 'down']['content_age_days'].mean()

print(f"Number 1: Total pages available for signal analysis: {total_pages:,}")
print(f"Number 2: Pages currently trending down: {declining_pages:,} ({decline_rate:.1f}%)")
print(f"Number 3: Avg age of declining pages: {avg_age_down:.0f} days vs Others: {avg_age_other:.0f} days")
print("\nConclusion: We have a massive dataset with a clear subset of declining pages. Comparing observable signals (like age) between these groups proves there is solid evidence to mine here.")
```
